```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)
```

```python
x = np.linspace(0, 2 * np.pi, 1000)
```

```python
amplitudes = [0.5, 1, 1.5, 2, 2.5]
frequencies = [1, 2, 3, 4, 5]
```

```python
plt.figure(figsize=(10, 6))
line_styles = ['-', '--', '-.', ':', (0, (5, 1))]

for A, w, style in zip(amplitudes, frequencies, line_styles):
    y = make_sine_wave(x, A, w)
    plt.plot(x, y, label=f"A={A}, w={w}", linestyle=style)

plt.title("Sine Waves with Varying Amplitudes and Frequencies")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

stars = pd.read_csv("stars.csv")
```

```python
print(stars.head())
print(f"Shape: {stars.shape}")
print("\nColumn names and data types:")
print(stars.dtypes)
print("Shape of the dataframe: (rows, columns):", stars.shape)
```

```python
avg_mass = stars['Mass (M☉)'].mean()
avg_temp = stars['Temperature (K)'].mean()
largest_radius_star = stars.loc[stars['Radius (R☉)'].idxmax()]
m_type_count = stars[stars['Spectral_Type'].str.startswith('M')].shape[0]

print(f"Average Mass: {avg_mass}")
print(f"Average Temperature: {avg_temp}")
print("\nStar with largest radius:")
print(largest_radius_star)
print(f"\nNumber of M-type stars: {m_type_count}")
```

```python
print(stars.columns)
```

```python
closest_stars = stars.sort_values(by='Distance (ly)').head(3)
print("\nThree closest stars:")
print(closest_stars)
```

```python
m_type_stars = stars[stars['Spectral_Type'].str.startswith('M')]
m_type_stars.to_csv("m_type_stars.csv", index=False)
```

```python
plt.figure(figsize=(8, 5))
plt.hist(stars['Temperature (K)'], bins=20, color='skyblue', edgecolor='black')
plt.title("Temperature Distribution of Stars")
plt.xlabel("Temperature (K)")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

penguins = sns.load_dataset("penguins")
import matplotlib.pyplot as plt  # if not already imported
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
sns.scatterplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='species', ax=ax1)
ax1.set_title("Bill Length vs. Flipper Length")
ax1.set_xlabel("Bill Length (mm)")
ax1.set_ylabel("Flipper Length (mm)")
ax1.legend()
sns.histplot(data=penguins, x='body_mass_g', bins=20, kde=True, ax=ax2)
ax2.set_title("Distribution of Penguin Body Mass")
ax2.set_xlabel("Body Mass (g)")
ax2.set_ylabel("Count")
plt.tight_layout()
plt.show()
```
